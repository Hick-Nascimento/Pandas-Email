```python
import pyautogui
import pyperclip
import time # PERMITI PAUSAR O CODIGO POR ALGUNS SEGUNDOS

pyautogui.PAUSE = 1

# Passo 1 - Entrar no sistema da empresa (drive)
pyautogui.hotkey("ctrl", "t") # PARA ABRIR UMA NOVA ABA
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter") # CLICAR NO BOTÃO DO TECLADO

time.sleep(5)
# Passo 2 - Navegar o sistema até encontrar a base de dados
pyautogui.click(x=382, y=247, clicks=2) # CLICAR COM O MOUSE
time.sleep(2)

# Passo 3 - Exportar no sistema até encontrar a base de dados
pyautogui.click(x=334, y=343)
pyautogui.click(x=1090, y=148)
pyautogui.click(x=895, y=511)
time.sleep(5)

```

```python
# Passo 4 - Calcular os indicadores (faturamento e quantidade de produtos vendidos)
# Se for usar outro editor de texto  baixe o pandas, numpy e openxl
import pandas as pd # as pd É UM APELIDO PARA O PANDAS PARA ABREVIAR

tabela = pd.read_excel(r"C://Users/Hick/Downloads/Vendas - Dez.xlsx") # "r" antes do caminho da pasta do pc é para levar diretamente a pasta
display(tabela)
faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum() # SELECIONAR UMA TABELA
```

```python
# Passo 5 - Enviar em email para a diretoria com od indicadores 
pyautogui.hotkey("ctrl", "t") # ABRE UMA ABA
pyperclip.copy("https://mail.google.com/mail/u/0/#inbox") # COPIA O LINK
pyautogui.hotkey("ctrl", "v") # COPIA O LINKteste@gmail.com
pyautogui.press("enter") # ENTRAR NO SITE
time.sleep(5)

# CLICAR NO BOTÃO ESCREVER
pyautogui.click(x=69, y=168)
time.sleep(7)

# ESCREVER DESTINATÁRIO (QUEM RECEBE O EMAIL)
pyautogui.write("teste@gmail.com")
pyautogui.press("tab")

#ESCREVER ASSUNTO
pyautogui.press("tab")
pyperclip.copy("Relatório de Vendas")

# ESCREVER O CORPO DO EMAIL
pyautogui.press("tab")

texto = f"""
Prezados, bom dia

O faturamento de ontem foi de: R${faturamento:,.2f}
A quantidade de produtos foi de: {quantidade:,}

Att
"""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

# ENVIAR O EMAIL
pyautogui.hotkey("ctrl", "enter")
```
